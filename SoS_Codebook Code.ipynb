```python
a = {}
for i in range(len(list_need)):
    print(list_need[i])
    temp_dic = {'code': {'Missing': 0}, 'label': {'Missing': 0}}
    dfcsv2.iloc[:, i] = dfcsv2.iloc[:, i].astype(str)  # Convert column i of dfcsv to strings
    for j in range(len(dfsav2.iloc[:, i])):
            if dfsav2.iloc[j, i] in temp_dic['label']:
                temp_dic['code'][dfcsv2.iloc[j, i]] += 1
                temp_dic['label'][dfsav2.iloc[j, i]] += 1
            else:
                temp_dic['code'][dfcsv2.iloc[j, i]] = 1
                temp_dic['label'][dfsav2.iloc[j, i]] = 1
    a[list_need[i]] = temp_dic
```

```python
def create_dataframe(key, data):
    code_data = data[key]['code']
    label_data = data[key]['label']
    data_list=[]
    for i in range(len(code_data)):
        dict={key:list(code_data.items())[i][1],'code':list(code_data.keys())[i],'label':list(label_data.keys())[i]}
        data_list.append(dict)
    for i in range(len(data_list)):
        try:
            data_list[i]['code'] = data_list[i]['code'].split('.')[0]
        except ValueError:
            pass
    return data_list
        
    
```

```python
for i in a.keys():
    data_list=create_dataframe(i,a)
    df=pd.DataFrame(data=data_list)
    df.columns=['count','code','label']
    df=df.set_index('count', drop=True)
    print(df)
    print()
```
